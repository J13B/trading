###The Trading Algorithm:

Here is the most basic version of our algorithm and will be refined later in the notebook. It works by comparing the prices between Google and Apple,( `self.pricea` for Apple and `self.priceg` for Google), and then buys / sells stock if the correlation of the stocks in the last 5 days is greater than (0.90). To determine whether to buy or sell Google stock, it checks the movement of Apple's stock. If Apple's stock went down yesterday relative to today, it buys Google;  if apple's stock went up it sells Google. It does this by evaluating price variables with a 1 period lagged price variable. At the daily level that period is 1 day; at the minute bar level that period is 1 minuet of tick data.

The stock does the oppiste for trading Apple's stock. In this way, the pricing algorithm checks the correlations to see if in fact the stocks have been following each other within expected tolerances and then trades each stock by the other's divergence. The algorithm to do so is shown below:


```python
class GOOG_AAPL_COR(TradingAlgorithm): # inherit from TradingAlgorithm
	"""
	Google Apple Correlation algorithm.

	An algorithm designed to exploit the correlation in the movement of Apple's and 
	Google's common stock. 
	"""
	def initialize(self):
		self.AUM = 1000000
        self.pricea = 0
        self.priceg = 0
		self.window_length = 5 
		self.corr_tolerance = 0.9
		self.max_notional = self.AUM + 0.1
		self.min_notional = self.AUM * (-1) 
		self.pricealag = 0
		self.priceglag = 0
		self.buy_orders = []
		self.sell_orders = [] 
		self.corr_values_aapl = []
		self.corr_values_goog = []

	def handle_data(self, data):

		pricea = data.AAPL.price
		priceg = data.GOOG.price   
        
		corr, pval = self.correlation(data, self.pricea, self.priceg)
		print 'Pearson Correlation: ' + str(corr)

		#notional = self.portfolio.positions[data.AAPL].amount * pricea
		#notional = notional + self.portfolio.positions[data.GOOG].amount * priceg 
        

		# need to include and optimization for # of shares
		if self.priceg < self.priceglag and corr > self.corr_tolerance:
			self.order('AAPL',-100)
			self.sell_orders.append((data.AAPL.datetime, pval))
			print "{dt}: Selling 100 AAPL shares.".format(dt = data.AAPL.datetime)
		elif self.priceg > self.priceglag and corr > self.corr_tolerance:
			self.order('AAPL',100)
			self.buy_orders.append((data.AAPL.datetime, pval))
			print "{dt}: Buying 100 AAPL shares.".format(dt = data.AAPL.datetime)

		if self.pricea < self.pricealag and  corr > self.corr_tolerance:
			self.order('GOOG',-100)
			self.sell_orders.append((data.GOOG.datetime, pval))
			print "{dt}: Selling 100 GOOG shares.".format(dt = data.GOOG.datetime)
		elif self.pricea > self.pricealag and corr > self.corr_tolerance:
			self.order('GOOG',100)
			self.buy_orders.append((data.GOOG.datetime, pval))  
			print "{dt}: Buying 100 GOOG shares.".format(dt = data.GOOG.datetime)

		# lag price variables 
		self.pricealag = data.AAPL.price
		self.priceglag = data.GOOG.price

	def correlation(self, data, pricea, priceg):
		self.corr_values_aapl.append(pricea)
		self.corr_values_goog.append(priceg)
		corr = pearsonr(self.corr_values_aapl[-self.window_length:], self.corr_values_goog[-self.window_length:]) # [0] only get the correltion
		return corr
```

###Lets simulate our algorithm on daily returns:

###Lets do better.

Lets clean the signal from the noise by adjusting Apple and Google's returns by the S&P returns. Through this process we are reducing the systematic noise in the movements of the two stock prices and getting closer to the actual signals of the two firm's movements. ###Let's incorporate this idea into our algorithm:

Here we will use the day's percent change instead of the adjusted closing value of the stock. Before we decide whether to buy or sell we will adjust the return by the return of the SPX in the same period. Then we will buy / sell the stock if they are moving within our correlation tolerance and if they are positive, (buy the stock), or negative, (sell the stock).

The resulting trading algorithm is shown below:

```python
class GOOG_AAPL_COR_V_SP500(TradingAlgorithm): # inherit from TradingAlgorithm
	"""
	Google Apple Correlation algorithm adjusted for movements in the S&P500.

	An algorithm designed to exploit the correlation in the movement of Apple's and 
	Google's common stock. 
	"""
	def initialize(self):
		self.AUM = 1000000
        self.pricea = 0
        self.priceg = 0
        self.pricespx = 0
		self.window_length = 5 
		self.corr_tolerance = 0.9
		self.max_notional = self.AUM + 0.1
		self.min_notional = self.AUM * (-1) 
		self.pricealag = 0
		self.priceglag = 0
		self.pricespxlag = 0
		self.buy_orders = []
		self.sell_orders = [] 
		self.corr_values_aapl = []
		self.corr_values_goog = []

	def handle_data(self, data):
		self.pricea = data.AAPL.price 
		self.priceg = data.GOOG.price  
		self.pricespx = data.SPX.price

		# find percent change and substract movements from SPX
		aapl_mov, goog_mov = self.spx_normalize(data, self.pricea, self.priceg, self.pricespx, self.pricealag, self.priceglag, self.pricespxlag)

		corr, pval = self.correlation(data, self.pricea, self.priceg)
		print 'Pearson Correlation: ' + str(corr)

		#notional = self.portfolio.positions[data.AAPL].amount * pricea
		#notional = notional + self.portfolio.positions[data.GOOG].amount * priceg 

		# need to include and optimization for # of shares
		if goog_mov < 0 and corr > self.corr_tolerance:
			self.order('AAPL',-100)
			self.sell_orders.append((data.AAPL.datetime, pval))
			print "{dt}: Selling 100 AAPL shares.".format(dt = data.AAPL.datetime)
		elif goog_mov > 0 and corr > self.corr_tolerance:
			self.order('AAPL',100)
			self.buy_orders.append((data.AAPL.datetime, pval))
			print "{dt}: Buying 100 AAPL shares.".format(dt = data.AAPL.datetime)

		if aapl_mov < 0 and  corr > self.corr_tolerance:
			self.order('GOOG',-100)
			self.sell_orders.append((data.GOOG.datetime, pval))
			print "{dt}: Selling 100 GOOG shares.".format(dt = data.GOOG.datetime)
		elif aapl_mov > 0 and corr > self.corr_tolerance:
			self.order('GOOG',100)
			self.buy_orders.append((data.GOOG.datetime, pval))  
			print "{dt}: Buying 100 GOOG shares.".format(dt = data.GOOG.datetime)

		# lag price variables 
		self.pricealag = data.AAPL.price
		self.priceglag = data.GOOG.price
		self.pricespxlag =  data.SPX.price

	def correlation(self, data, pricea, priceg):
		self.corr_values_aapl.append(self.pricea)
		self.corr_values_goog.append(self.priceg)
		corr = pearsonr(self.corr_values_aapl[-self.window_length:], self.corr_values_goog[-self.window_length:]) # [0] only get the correltion
		return corr

	def per_change(self, data, begin, end):
		percent_change = ((end - begin)/begin)
		return percent_change
		
	def spx_normalize(self, data, pricea, priceg, pricespx, pricealag, priceglag, pricespxlag):
		spx_change = self.per_change(data, pricespx, self.pricespxlag)
		aapl_change = (self.per_change(data, pricea, self.pricealag) - spx_change)
		goog_change = (self.per_change(data, priceg, self.priceglag) - spx_change)
		return (aapl_change, goog_change)
        
```

```python
GOOG_AAPL_COR_V_SP500 = GOOG_AAPL_COR_V_SP500()
results_GOOG_AAPL_COR_V_SP500 = GOOG_AAPL_COR_V_SP500.run(data)
```

```python
class GOOG_AAPL_COR_V_SP500_Optimized(TradingAlgorithm): # inherit from TradingAlgorithm
    """
    Google Apple Correlation algorithm adjusted for movements in the S&P500.

    An algorithm designed to exploit the correlation in the movement of Apple's and 
    Google's common stock. 
    """
    def initialize(self):
        self.AUM = 1000000
        self.pricea = 0
        self.priceg = 0
        self.pricespx = 0
        self.trade_size_aapl = 0
        self.trade_size_goog = 0 
        self.window_length = 5 
        self.corr_tolerance = 0.9
        self.max_notional = self.AUM + 0.1
        self.min_notional = self.AUM * (-1) 
        self.pricealag = 0
        self.priceglag = 0
        self.pricespxlag = 0
        self.buy_orders = []
        self.sell_orders = [] 
        self.corr_values_aapl = []
        self.corr_values_goog = []

    def handle_data(self, data):
        self.pricea = data.AAPL.price 
        self.priceg = data.GOOG.price  
        self.pricespx = data.SPX.price
        
        self.trade_size_aapl = round(((self.AUM * .75) / float(self.pricea)))
        self.trade_size_goog = round(((self.AUM * .75) / float(self.priceg)))

        # find percent change and substract movements from SPX
        aapl_mov, goog_mov = self.spx_normalize(data, self.pricea, self.priceg, self.pricespx, self.pricealag, self.priceglag, self.pricespxlag)

        corr, pval = self.correlation(data, self.pricea, self.priceg)
        print 'Pearson Correlation: ' + str(corr)

        #notional = self.portfolio.positions[data.AAPL].amount * pricea
        #notional = notional + self.portfolio.positions[data.GOOG].amount * priceg 

        if goog_mov < 0 and corr > self.corr_tolerance:
            self.order('AAPL',-(self.trade_size_aapl))
            self.sell_orders.append((data.AAPL.datetime, pval))
            print "{dt}: Selling {ts} AAPL shares.".format(dt = data.AAPL.datetime, ts = self.trade_size_aapl)
        elif goog_mov > 0 and corr > self.corr_tolerance:
            self.order('AAPL',(self.trade_size_aapl))
            self.buy_orders.append((data.AAPL.datetime, pval))
            print "{dt}: Buying {ts} AAPL shares.".format(dt = data.AAPL.datetime, ts = self.trade_size_aapl)

        if aapl_mov < 0 and  corr > self.corr_tolerance:
            self.order('GOOG',-(self.trade_size_goog))
            self.sell_orders.append((data.GOOG.datetime, pval))
            print "{dt}: Selling {ts} GOOG shares.".format(dt = data.GOOG.datetime, ts = self.trade_size_goog)
        elif aapl_mov > 0 and corr > self.corr_tolerance:
            self.order('GOOG', self.trade_size_goog)
            self.buy_orders.append((data.GOOG.datetime, pval))  
            print "{dt}: Buying {ts} GOOG shares.".format(dt = data.GOOG.datetime, ts = self.trade_size_goog)

        # lag price variables 
        self.pricealag = data.AAPL.price
        self.priceglag = data.GOOG.price
        self.pricespxlag =  data.SPX.price

    def correlation(self, data, pricea, priceg):
        self.corr_values_aapl.append(pricea)
        self.corr_values_goog.append(priceg)
        corr = pearsonr(self.corr_values_aapl[-self.window_length:], self.corr_values_goog[-self.window_length:]) # [0] only get the correltion
        return corr

    def per_change(self, data, begin, end):
        percent_change = ((end - begin)/begin)
        return percent_change
        
    def spx_normalize(self, data, pricea, priceg, pricespx, pricealag, priceglag, pricespxlag):
        spx_change = self.per_change(data, pricespx, self.pricespxlag)
        aapl_change = (self.per_change(data, pricea, self.pricealag) - spx_change)
        goog_change = (self.per_change(data, priceg, self.priceglag) - spx_change)
        return (aapl_change, goog_change)
        
```
